```python
import pandas as pd
import os 
```

```python
base=pd.read_csv("../files/input/solicitudes_de_credito.csv", sep=";")
base.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
base.info()
```

```python
for col in base.columns.to_list():
    print(base[col].value_counts())
```

```python
base_1=base.copy()
base_1=base_1.dropna()
base_1=base_1.drop_duplicates()
base_1.info()
```

```python
base_1=base_1.copy()
base_1["sexo"]=(base_1["sexo"].str.strip()
                .str.lower())
base_1["sexo"].value_counts()
```

```python
base_1["tipo_de_emprendimiento"]=(base_1["tipo_de_emprendimiento"].str.strip()
                                 .str.lower())
base_1["tipo_de_emprendimiento"].value_counts()
```

```python
display(base_1["idea_negocio"].unique().tolist())
display(base_1["idea_negocio"].value_counts())
```

```python
base_1["idea_negocio"]=(base_1["idea_negocio"].str.replace("_"," ")
                        .str.replace("-"," ")
                        .str.strip()
                        .str.lower()
                        .str.replace(" ","_")
                        .str.replace("-","_"))
base_1["idea_negocio"].value_counts()
```

```python
base_1["idea_negocio"].unique()
```

```python
base_1["barrio"].value_counts()
```

```python
base_1["barrio"].unique()
```

```python
base_1["barrio"]=(base_1["barrio"].str.replace("_"," ")
                        .str.replace("-"," ")
                        .str.strip()
                        .str.lower()
                        .str.replace(" ","_")
                        .str.replace("-","_"))
base_1["barrio"].value_counts()
```

```python
valores=base_1["barrio"].unique()
valores=valores.tolist()
valores.sort()
valores
```

```python
base_1["estrato"]=base_1["estrato"].apply(lambda x: str(int(x)))
```

```python
base_1["comuna_ciudadano"]=base_1["comuna_ciudadano"].apply (lambda x: 
                                                             str(int(x/10)) if x>16 else str(int(x)))
base_1["comuna_ciudadano"].value_counts()   
```

```python
fechas=base_1["fecha_de_beneficio"].unique()
fechas=fechas.tolist()
fechas.sort()
fechas
for i in fechas:
    print(i)
```

```python
base_1["fecha_de_beneficio"]=(base_1["fecha_de_beneficio"].str.strip()
                              .str.replace("/","-"))

base_1["fecha_de_beneficio"]=base_1["fecha_de_beneficio"].str.replace(
    r'^(\d{1,2})-(\d{1,2})-(\d{4})$',  
    r'\3-\2-\1',                       
    regex=True)

base_1["fecha_de_beneficio"]=pd.to_datetime(base_1["fecha_de_beneficio"], errors='coerce')
```

```python
base_1["fecha_de_beneficio"].value_counts()
base_1["fecha_de_beneficio"].isnull().sum()
```

```python
base_1.info()
```

```python
base_1["monto_del_credito"]=(base_1["monto_del_credito"]
                            .str.replace("$","")
                            .str.replace(",","")
                            .str.strip())
base_1["monto_del_credito"]=base_1["monto_del_credito"].apply(lambda x: float(x))
```

```python
base_1["monto_del_credito"].value_counts()
```

```python
base_1["línea_credito"].value_counts()
```

```python
base_1["línea_credito"]=(base_1["línea_credito"].str.lower()
                         .str.replace(r'[-_]'," ", regex=True)
                         )
```

```python
base_1["línea_credito"].value_counts()
```
